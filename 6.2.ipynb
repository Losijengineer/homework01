# 图像网格化

```python
from PIL import Image, ImageDraw,ImageFont
import matplotlib.pyplot as plt
font = ImageFont.truetype("consola.ttf", 40, encoding="unic")#设置字体
step=10
img = Image.open('1.jpg')
w,h=img.size #参数宽和高
print(w,h)
draw =ImageDraw.Draw(img)
for col in range(0,w,w//step):
    draw.line((col, 0, col, h), 'black',width=5)
for row in range(0,h,h//step):
    draw.line((0, row, w, row), 'black',width=5)
psum=step**2
for num in range(psum):
    x1=(num%step)*(w//step)
    y1=(num//step)*(h//step)
    draw.text((x1, y1), str(num), 'black', font)
plt.figure(figsize=(100,100))
plt.axis('off')
plt.imshow(img)
plt.show()
```
